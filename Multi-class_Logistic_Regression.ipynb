```python
#import the packages and load the dataset
import numpy as np
import pandas as pd
df=pd.read_csv('mnist_train_small.csv')
```

```python
#split df into feature and label
y,x=np.hsplit(df,[1])
print(y.head())
print(x.head())
```

```python
#normalize the feature variables
X=x.copy()
maximum=X.max().max()
print(maximum)
X=X.div(maximum)
X=X.astype('float64')
print (X.sample(5))
y /= 10
print(y.head())
```

```python
#Add the bias column to the feature dataframe 'X'
X=pd.concat([pd.Series(1.0,index=X.index,name='00'),X],axis=1)
print(X.head())
row=X.shape[0]
print(row)
col=X.shape[1]
print(col)
```

```python
#check out how many unique classes are present
y=y.iloc[:,0]
print(y.unique())
u=y.unique()
clas=len(u)
print(clas)
```

```python
#convert the dataframes into numpy arrays
y=y.to_numpy()
X=X.to_numpy()
```

```python
from numpy.random import rand
w = np.array(rand(1, col)) #initializing weights to random values
print(w.shape, w)
```

```python
import random
w0 = random.random()
print(w0)
```

```python
epoch = 50
eta = 0.1
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x/255))
def sigmoid_grad(x):
    return x*(1-x)
```

```python
def train(X, y):
    # wji = eta* delj * xji, where delj = oj * (1-oj) * (tj-oj)
    global row, w, col
    for j in range(epoch):
        for i in range(row):
            r = X[i:i+1]; t = y[i:i+1]
            oi = np.dot(w,r.T)
            #print(oi)
            oi = w0 + oi.item() #take 1 row at a time
            oi = sigmoid(oi)
            #print("oi=",oi,oi.shape)
            deli = sigmoid_grad(oi) * (t-oi)
            #print("deli=",deli)
            #print(w.shape, r.shape)
            w = w + np.multiply(r,deli) * eta
            #print("w=",w)
```

```python
train(X, y)
print(w)
```
